```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# Zebra desenini oluşturmak için fonksiyon
def zebra_pattern(ax, num_stripes=10, stripe_width=0.1):
    # Deseni çizmek için siyah ve beyaz şeritler oluştur
    for i in range(num_stripes):
        # Alternatif olarak siyah ve beyaz şeritler
        color = 'black' if i % 2 == 0 else 'white'
        
        # Şeritlerin yatayda paralel olarak yerleştirilmesi
        ax.add_patch(plt.Rectangle((0, i * stripe_width), 1, stripe_width, color=color))

# Zebra başı çizimi (çok basit bir versiyon)
def draw_zebra_head(ax):
    # Gözler
    ax.add_patch(Ellipse((0.3, 0.8), 0.1, 0.15, color='black'))
    ax.add_patch(Ellipse((0.7, 0.8), 0.1, 0.15, color='black'))
    
    # Burun
    ax.add_patch(Ellipse((0.5, 0.5), 0.2, 0.25, color='black'))
    
    # Ağız
    ax.plot([0.4, 0.5, 0.6], [0.4, 0.3, 0.4], color='black', lw=2)

# Zebra deseni ve başını çiz
fig, ax = plt.subplots(figsize=(6, 6))

# Zebra desenini çiz
zebra_pattern(ax)

# Zebra başını ekle
draw_zebra_head(ax)

# Eksenleri kaldır
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')

plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Beşiktaş logosunun basit bir versiyonunu çizen fonksiyon
def besiktas_logo():
    fig, ax = plt.subplots(figsize=(6, 6))

    # Daire (logo dış çerçevesi)
    circle = plt.Circle((0.5, 0.5), 0.45, color='black', lw=5, fill=False)
    ax.add_patch(circle)

    # Beşiktaş'ın renkleri olan siyah ve beyaz şeritler
    # Üst ve alt şeritler
    ax.add_patch(plt.Rectangle((0, 0.5), 1, 0.1, color='black'))
    ax.add_patch(plt.Rectangle((0, 0), 1, -0.1, color='white'))

    # Sağ ve sol şeritler
    ax.add_patch(plt.Rectangle((0, 0.35), 0.1, 0.3, color='black'))
    ax.add_patch(plt.Rectangle((0.9, 0.35), 0.1, 0.3, color='white'))

    # İç daire (logo iç kısmı)
    inner_circle = plt.Circle((0.5, 0.5), 0.3, color='white', lw=3)
    ax.add_patch(inner_circle)

    # "1903" yılı yazısı (logodaki tarih)
    ax.text(0.5, 0.5, '1903', fontsize=30, ha='center', va='center', color='black', fontweight='bold')

    # Eksenleri kaldır
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal')
    ax.axis('off')

    plt.show()

# Beşiktaş logosunu çiz
besiktas_logo()
```

```python
import matplotlib.pyplot as plt

# Almanya bayrağını çizen fonksiyon
def germany_flag():
    fig, ax = plt.subplots(figsize=(6, 4))

    # Bayrağın üç yatay şeridi: siyah, kırmızı ve sarı
    ax.add_patch(plt.Rectangle((0, 0.67), 1, 0.33, color='black'))  # Siyah şerit
    ax.add_patch(plt.Rectangle((0, 0.34), 1, 0.33, color='red'))    # Kırmızı şerit
    ax.add_patch(plt.Rectangle((0, 0), 1, 0.34, color='gold'))      # Sarı şerit

    # Eksenleri kaldır
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal')
    ax.axis('off')

    plt.show()

germany_flag()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def rainbow():
    # Gökkuşağının yarıçapları
    radii = np.linspace(0.1, 1, 7)

    # Gökkuşağının renkleri (7 renkli sıralama)
    colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']

    # Grafik alanını ayarlıyoruz
    fig, ax = plt.subplots(figsize=(8, 6))

    # Her bir renk için yarım daireyi çiziyoruz
    for i, radius in enumerate(radii):
        # Her renk için büyük bir daire çizeceğiz
        circle = plt.Circle((0.5, 0), radius, color=colors[i], clip_on=True, lw=0)
        ax.add_artist(circle)

    # Yarım daireyi kesmek için bir kırmızı daire ekleyelim
    ax.add_patch(plt.Circle((0.5, 0), 1.05, color='white'))

    # Grafik alanını ayarlıyoruz
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal')
    ax.axis('off')  # Eksenleri kaldırıyoruz

    plt.show()

# Gökkuşağını çiz
rainbow()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def rainbow():

    colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']


    fig, ax = plt.subplots(figsize=(8, 6))

    for i, color in enumerate(colors):
        ax.axhline(y=1 - (i * 0.14), color=color, linewidth=8)


    ax.axis('off')
    
    plt.show()

rainbow()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def rainbow():
    
    colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']
    
    
    num_bows = len(colors)  
    angles = np.linspace(0, np.pi, 100)  
    radii = np.linspace(0.1, 1, num_bows)  

    fig, ax = plt.subplots(figsize=(8, 6))


    for i, color in enumerate(colors):
  
        radius = radii[i]
        
        x = radius * np.cos(angles)
        y = radius * np.sin(angles)
        
       
        ax.plot(x + 0.5, y + 0.2, color=color, lw=6)  # X ve Y'yi kaydırarak yayları üst üste çiziyoruz

   
    ax.axis('off')
    
print()

rainbow()
```
